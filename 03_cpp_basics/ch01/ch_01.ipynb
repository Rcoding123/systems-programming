#### Variable assignment and initialization

**Variable Assignment**<br>
Give a value to a variable using the *=* operator <br>
This process is called assignment (using the assignment operator)<br>
Known as **copy-assignment**

```c++
int width; // definition of width (int)
width = 10; // assignment of value 10 into width
```

#### Variable Initialization
Giving a value to the variable at its definition<br>
There are different types of initializers:
- int a;      (no initializer)
- int b = 5;  (copy-initialization)
- int c(6);   (direct-initialization)
- int d {10}; (direct-list-initialization)
- int e {};   (value-initialization)

```c++
int width {5}; // define variable width and initialize with 5
```

### Introduction to iostream: cout, cin, endl

The input/output library is part of the C++ standard library<br>
This deals with the functionality of getting input from the keyboard and output to the console
<br> Uses #include iostream

**std::cout**<br>
Sends data to the console: cout (character out)<br>
Uses the (<<) operator - insertion operator to send the characters to the console to be printed<br>
The << operator can be used to chain multiple pieces of output

```c++
std::cout << "Hello" << " world!" << '\n'; // How does work??? -> what allows the chaining to happen
```

**std::ostream& operator<<(std::ostream& os, const T& value)** <br>
This is the signature of the function for (<<)<br>
Because it returns the same stream (by reference) -> the expression expands onto itself<br>
std::cout << a << b parses as: <br>
((std::cout << a) << b) << c; <br>
return is std::cout << b <br>
return is std::cout << c;<br>

It always prints first and then returns

**std::cin** <br>
Reads input from the keyboard: cin (character in) <br>
Uses the (>>) operator - extraction opertor to get from the stream and put into the variable<br>
The >> operator can be used to chain multipe pieces of input

```c++
int x{};
std::cin >> x;
std::cout << "Value of x: " << x << '\n
```
